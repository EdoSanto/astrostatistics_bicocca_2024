```python
import numpy as np
from scipy.stats import poisson
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
plt.rcParams['figure.figsize'] = [8, 8]
```

```python
deaths = np.zeros(shape = 200) 
for i in range(109, 174):
    deaths[i] = 1

for i in range(174, 196):
    deaths[i] = 2
    
for i in range(196, 199):
    deaths[i] = 3
    
deaths[199] = 4

plt.hist(deaths, bins = [-0.5,0.5,1.5,2.5,3.5,4.5], density= True)

dist = poisson(np.mean(deaths))

x = np.arange(-1, 10)
plt.plot(x, dist.pmf(x), color='black')
    
plt.xlim(0, 5)
plt.ylim(0, 0.6)

plt.xlabel('$x$')
plt.ylabel(r'$p(x|\mu)$')
plt.title('Poisson Distribution')

plt.show()
```
