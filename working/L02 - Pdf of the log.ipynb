```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100000
x = np.random.uniform(0.1, 10, N)
y = np.log10(x)

def P_y(y):
    return np.log(10)/9.9*pow(10,y) # pdf of the log

Y = np.sort(y)
fig, axs = plt.subplots(2, 1, figsize=(5,10))
axs[0].hist(x, 100, density = True)
axs[1].hist(y, 100, density = True)
axs[1].plot(Y, P_y(Y))

print(np.log10(np.mean(x)), np.mean(y))
print(np.log10(np.median(x)), np.median(y))
```

While mean is affected by the scale of the sample values, the median only depends on the ordering of samples, and monotonic transformations do not change the ordering of samples.
